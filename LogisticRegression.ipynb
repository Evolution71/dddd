```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('insurance.csv',low_memory=False)
df
```

```python
plt.scatter(df.age,df.b_insurance,marker='+',color='red')## we are plotting to check the distribution of the dataset
plt.ylabel('BOUGHT INSURANCE')
plt.xlabel('AGE')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['age']],df.b_insurance,test_size=0.2)  ##here we are splitting our data into x and y
```

```python
x_test  ## here is our testing data
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)  ###here my data accuracy is 1 that means is 100% accurate to predict
## our result was 1 because we didnt have large data sample ,there was no loss due to error,but wen we have larger 
# dataset we will have an accutate value less than 1
```

```python
model.predict_proba(x_test)
```
