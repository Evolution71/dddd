```python
days = np.arange(1,25)
month = ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
np.random.seed(5)
sales = np.random.randint(1000000,100000000,288)
sales = sales.reshape(24,12)
df = pd.DataFrame(data = sales, columns=month, index=days)
df.head()
```

```python
df['Total_Sales'] = df['JAN'] + df['FEB'] + df['MAR'] + df['APR'] + df['MAY'] + df['JUN'] + df['JUL'] + df['AUG'] + df['SEP'] + df['OCT'] + df['NOV'] + df['DEC']
df.head()
```
