```python
## But instead of writing a code to check for our outliers everytime,lets create a fuction so that we can call the function when checking for outliers
def outliers_detector(movie,col):
    mean=movie[col].mean()
    std=movie[col].std()

    outliers=[]
    for data in movie[col]:
        z_score=(data-mean)/std
        if z_score > 3 or z_score < -3: ## value 3 is best used to detect an outlier in a dataset using Z-score
            outliers.append(data)
    return outliers
```

```python
##Now lets call the function we created for outlier detection in rating
outliers_detector(movie,'Rating')
```

```python
##Find the limit to use and detect outliers
upper_limit= mean + (3*std)
lower_limit=mean - (3*std)
print(f'The upper limit is {upper_limit} and The lower limit is {lower_limit}')
```

```python
import re
movie['Duration']=movie['Duration'].apply(lambda x: re.sub(r'\d\:hms','',str(x)))
movie['Duration'].unique()
##Here we noticed a link in our Duration column,so we need to remove them
#movie_df['Duration']=movie_df['Duration'].replace(to_replace=r'https?:\/\/.*[\r\n]*' ,value='' ,regex=True)
#print(movie_df)
```

```python
##Clean the link and other invalid values
def clean_duration(text):
    pattern = r'(\d+\s*h\s*\d+\s*m)'
    match = re.search(pattern, text)
    return match.group(0).replace(' ', '')if match else ''
movie['Duration']=movie['Duration'].apply(clean_duration)
movie['Duration'].unique()
```

```python
##Here the links are removed successfully
movie['Duration'].unique()
```

```python
def duration_to_minute(text):
    match = re.match(r'(?:(\d+)h)?(?:(\d+)m)?', text)

    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0
        return hours * 60 + minutes
    return 0
movie['Duration']=movie['Duration'].apply(duration_to_minute)
duration_to_minute
movie['Duration'].unique()
```

```python
def categorize_duration(minutes):
    if minutes < 80:
        return "Short"
    elif minutes < 250:
        return 'Medium'
    else:
        return 'Long'
movie['Duration_Category'] = movie['Duration'].apply(categorize_duration) 
movie.head()
```
