```python
## handli g of missing values
## filling missing value using forward fill
df1 = df.copy()
df1.fillna(method='ffill',inplace = True)
df1.tail()
```

```python
## handling of missing values using backward fill
df1 = df.copy()
df1.fillna(method='bfill',inplace = True)
df1.tail()
```

```python
## filling using statiscal analysis
df1 = df.copy()
df1['Units'].fillna(df['Units'].median(),inplace = True)
df1.tail()
```

```python
## using replacement method
df1 = df.copy()
df1['Units'].replace({np.NaN:0},inplace=True)
df1.tail()
```

```python
### dropping of values 
df.dropna(inplace=True)
```
