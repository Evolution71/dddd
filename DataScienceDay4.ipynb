```python
rate = pd.read_csv('exchange07082024.csv')
rate.head()
```

```python
rate.reset_index(inplace=True)
rate.head()
```

```python
rate.columns
```

```python
rate.rename({'index':'Rate Date', 'Rate Date':'Currency', 'Currency':'Rate Year', 'Rate Year':'Rate Month', 'Rate Month':'Buying Rate',
       'Buying Rate':'Central Rate', 'Central Rate':'Selling rate', 'Selling Rate':'Nill'}, axis =1, inplace=True)
rate.head()
```

```python
rate.drop('Nill',axis = 1,inplace=True)
rate.head()
```

```python
rate['Currency'].duplicated()
```

```python
rate['Currency'].unique()
```

```python
rate['Currency'].replace({'POUND STERLING':'POUNDS STERLING', 'DANISH KRONER':'DANISH KRONA',
       'JAPANESE YEN':'YEN', 'US DOLLAR ':'US DOLLAR', 'YEN ':'YEN', 'CFA ':'CFA', 'WAUA ':'WAUA', 'SDR ':'SDR',
       'EURO\t':'EURO', 'EURO ':'EURO', 'RIYAL ':'RIYAL', 'SWISS FRANC\t':'SWISS FRANC', 'SDR\t':'SDR'},inplace =True)
rate['Currency'].unique()
```

```python
## merging of two tables
usd = rate[rate['Currency'] == 'US DOLLAR']
euro = rate[rate['Currency'] == 'EURO']
pounds = rate[rate['Currency'] == 'POUNDS STERLING']
waua = rate[rate['Currency'] == 'WAUA']
```

```python
usd.merge(euro, how='outer')
```

```python
all = pd.concat([euro, usd,waua,pounds],axis = 0)
all = all.sample(frac=1)
all
```

```python
## save to csv and excel
all.to_csv('Rate.csv',index=False)
```

```python
all.to_excel('Rate.xlsx',index=False, sheet_name='RATE')
```

```python
with pd.ExcelWriter('Currency.xlsx',engine='xlsxwriter') as writer:
    usd.to_excel(writer, sheet_name='USD', index=False,engine='xlsxwriter')
    euro.to_excel(writer, sheet_name='EURO', index=False,engine='xlsxwriter')
    pounds.to_excel(writer, sheet_name='POUNDS', index=False,engine='xlsxwriter')
    waua.to_excel(writer, sheet_name='WAUA', index=False,engine='xlsxwriter')
```
