```python
##Create a line plot to visualize the trends of a stock's closing prices
## I installed yfinance to be able to download the stock table
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from matplotlib import style
import yfinance as yf
import seaborn as sns
stocks = "AAPL"
data = yf.download(stocks)
data.head()
```

```python
##Because i want to visualize 2023 to 2024 i will have to put my starting date and my ending date
stock = yf.download(stocks, start ='2023-01-01', end='2024-05-06')
dataa = stock.loc[:, "Adj Close"].copy()
dataa.head()
```

```python

dataa.plot(figsize=(17,8), fontsize=18)
plt.style.use("ggplot")
plt.show()
```
